# student data administration project

###  this project gives the user ability to input student's data and store it in text file starting with selecting one of the following available options:
* first one is to input the student data as the form of first_name last_name term gpa regestration_number
* second option is to present all the stored data
* third option is to search for student by his/her regestration number
* fourth option is to remove any student from file
* fifth option is to update existing data in text file
* last option is to sort data forward or reversely

### 8 functions are used to increase the code effectviness & facilitate its duty
### function names comes as following:
* enter_std() => this function stores data in RAM it make sure that the data entered and stored in the right format
* r_std() => this function is used to transfer the existing data in the text file to RAM
* w_std() => this function is used to transfer data from RAM to text file using one of the options: overwrite, append or empty the file
* print_all(stds) => present all student data stored in RAM
* search(stds, reg) => searching for specific student in RAM by his/her regestration number
* remove(stds, reg) => removing student data from both RAM and text file
* update(stds, reg) => updating student data in both RAM and text file
* sort_std(stds, opt) => sorting data by first name in RAM

```python
 def enter_std(): 
    name = input("enter FN LN TERM GPA REG_NU: ").lower()
    if len(name.split()) == 5:
        return(name)
    else:
        print("you may have entered more or less than the required data.. please enter data in form of FN LN TERM GPA REG_NU")
        return("wrong")

def r_std():
    hand = open("test.txt", "r")
    stds = hand.read()
    stds = stds.split("\n")
    if "" in stds:
        stds.remove("")
    hand.close()
    return(stds)

def w_std(name = "", opt = "a", empty = 0):
    hand = open("test.txt", opt)
    if empty == 0:
        hand.write(name + "\n")
    hand.close()

def print_all(stds):
    for name in stds:
        name = name.split()
        print("FN: ", name[0], "LN: ", name[1], "TERM: ", name[2], "GPA: ", name[3], "REG_NU: ", name[4], end = "\n\n")

def search(reg, stds):
    for name in stds:
        if reg == name.split()[4]:
            return(name)
    else:
        return("not found")

def remove(reg, stds):
    name = search(reg, stds) 
    if name != "not found":
        stds.remove(name) 
        w_std(opt = "w", empty = 1) 
        for student in stds: 
            w_std(student)
        return(name) 
    else:
        return("not found")

def update(reg, stds):
    name = search(reg, stds)
    if name != "not found":
        stds.remove(name)
        w_std(opt = "w", empty = 10)
        new_data = input("please enter the new data in the form of FN LN TERM GPA REG_NU ")
        stds.append(new_data)
        for student in stds:
            w_std(student)
        return(name)
    else:
        return("not found")

def sort_std(stds, opt):
    if opt == 1:
        return(sorted(stds))
    else:
        return(sorted(stds, reverse = True))
    
while True:
    opt = input("1-enter new std 2-print all 3-search 4-remove 5-update 6-sort or done to exit: ").lower()
    if opt == "done":
        break
    elif opt == "1":
        name = enter_std()
        if name != "wrong":
            w_std(name)
    elif opt == "2":
        stds = r_std()
        print_all(stds)
    elif opt == "3":
        stds = r_std()
        reg = input("please enter reg_number you want to search for: ")
        name = search(reg, stds)
        print("student that you searched for : ", name)
    elif opt == "4":
        stds = r_std()
        reg = input("please enter regestration number you want to remove ")
        removed_name = remove(reg, stds)
        if removed_name != "not found":
            print("student: ", removed_name, "is removed")
        else:
            print("student is not found")
    elif opt == "5":
        stds = r_std()
        reg = input("please enter regestration number you want to update ")
        updated_name = update(reg, stds)
        if updated_name != "not found":
            print("student data has been succesfully updated from", updated_name, "to", stds[-1])
        else:
            print("student is not found")
    elif opt == "6":
        stds = r_std()
        opt = int(input("please type 1 to sort FW and type 2 to sort REV ")) # any number input except 1 is REV
        sorted_names = sort_std(stds, opt)
        print(sorted_names)
```
